```
import numpy as np
import skfuzzy as sk
import matplotlib.pyplot as plt
```

```
x_qual = np.arange(0, 10.1, .1)
```

```
qual_lo = sk.trimf(x_qual, [0,5,10])
plt.plot(x_qual, qual_lo)
```

```
qual_lo = sk.trapmf(x_qual, [2, 5, 7, 8])
plt.plot(x_qual, qual_lo)
```

```
qual_lo = sk.gaussmf(x_qual, np.mean(x_qual), 2)
plt.plot(x_qual, qual_lo)
```

```
qual_lo = sk.gbellmf(x_qual, 3, 3, 5)
plt.plot(x_qual, qual_lo)
```

```
qual_lo = sk.sigmf(x_qual, 5,-2)
plt.plot(x_qual, qual_lo)
```

```
#Defining the Numpy array for Tip Quality
x_qual = np.arange(0, 11, 1)
#Defining the Numpy array for two membership functions (Triangular)
qual_lo = sk.trimf(x_qual, [0, 0, 5])
qual_md = sk.trimf(x_qual, [0, 5, 10])
#Finding the Maximum (Fuzzy Or)
union = sk.fuzzy_or(x_qual,qual_lo,x_qual,qual_md)
```

```
plt.plot(x_qual, qual_lo, label='low')
plt.plot(x_qual, qual_md, label='md')
# plt.plot(union[0], union[1],'-',  label='union', alpha=0.5)
plt.fill_between(union[0], union[1], alpha=.2, label='union')
plt.legend()
plt.show()
```

```
def visualize_operation(univ1, mf1, univ2, mf2, operation='union'):
    if operation == 'union':
        res_set = sk.fuzzy_or(univ1,mf1,univ2,mf2)
    if operation == 'intersection':
        res_set = sk.fuzzy_and(univ1,mf1,univ2,mf2)
    if operation == 'difference':
        res_set = sk.fuzzy_sub(univ1,mf1,univ2,mf2)

    plt.plot(univ1, mf1, label='set 1')
    plt.plot(univ2, mf2, label='set 2')
    # plt.plot(union[0], union[1],'-',  label='union', alpha=0.5)
    plt.fill_between(res_set[0], res_set[1], alpha=.2, label=operation)
    plt.legend()
    plt.title(operation)
    plt.xlabel('Universe of discourse')
    plt.ylabel('membership')
    plt.show()
```

```
#Defining the Numpy array for Tip Quality
x_qual = np.arange(0, 11, 1)
#Defining the Numpy array for two membership functions (Triangular)
qual_lo = sk.trimf(x_qual, [0, 0, 5])
qual_md = sk.trimf(x_qual, [0, 5, 10])

visualize_operation(x_qual,qual_lo,x_qual,qual_md, 'union')
visualize_operation(x_qual,qual_lo,x_qual,qual_md, 'intersection')
visualize_operation(x_qual,qual_lo,x_qual,qual_md, 'difference')
```
